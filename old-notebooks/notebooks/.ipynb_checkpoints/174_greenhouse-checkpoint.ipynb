```python
# Define Constants
SIGMA = 5.67 * 10**-8 # Stephan Boltzman constant 
F_s = 1028 #Solar Constant W/m^2
A = 0.7 # Albedo, fraction

def temp(f):
    T_o = ((F_s*(1-A))/((4-2*f)*SIGMA)) ** (1/4)
    return T_o

f_vals = [0, 0.77, 1]

for f in f_vals:
    print(f"Surface temp at f = {f} is %.2f Kelvin" % temp(f))
```
