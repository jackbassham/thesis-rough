```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA
import os
```

```python
PATH_SOURCE = "/home/jbassham/jack/data/weddell/1992_2020"
PATH_DEST = PATH_SOURCE
HEM = "sh" # nh or sh
START_YEAR = 1992
END_YEAR = 2020
```

```python
# Middle Weddell indices

iy = 40
ix = 200
```

```python
# Extract variables, slicing at single point 

fnam = "motion_ppv4_latlon_sh_1992_2020.npz"

data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle=True)
ui = data['u'][:,iy,ix] # zonal ice velocity
vi = data['v'][:,iy,ix] # meridional ice velocity
r = data['error'][:,iy,ix]
lat = data['lat']
lon = data['lon']
time = data['time']
```

```python
fnam = "con_nimbus7_latlon_sh_1992_2020.npz"
data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle=True)
ic = data['ic'][:,iy,ix] # zonal ice velocity
```

```python
fnam = "wind_JRA55_latlon_global_1992_2020.npz"
data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle=True)
uw = data['u'][:,iy,ix]
vw = data['v'][:,iy,ix]
```

```python
# Shift present day parameters forward one day, for one point Middle Weddell
uit = ui[1:]
vit = vi[1:]
uwt = uw[1:]
vwt = vw[1:]
ict = ic[1:]
tt = time[1:]
rt = r[1:]

# Remove last day from previous day parameters
icy = ic[:-1]
ty = time[:-1]
```

```python
# # Mask ice concentration

# ic_raw = np.round(ic * 250)

# # Mask 
# # 251 pole hole
# # 252 unused data
# # 253 coastline
# # 254 land
# icy = np.where((ic_raw == 251) | (ic_raw == 252) | (ic_raw == 253) | (ic_raw == 254), np.nan, icy)

# Filter points with ice error greater than 999 and 0 concentration ice motion
# Create mask
mask = (ict == 0) | (np.isnan(ict)) | (rt >= 999) | (rt <= -999)

# NaN out points with above condition
uit = np.where(mask, np.nan, uit) 
uvt = np.where(mask, np.nan, vit)
rt = np.where(mask, np.nan, rt)
uwt = np.where(mask, np.nan, uwt)
vwt = np.where(mask, np.nan, vwt)

# NaN out yesterday's ice concentration for consitency
icy = np.where(mask, np.nan, icy)
```

```python
# Find indices where ice motion is NaN
inan = np.where(np.isnan(uit) | np.isnan(vit))

# Create NaN mask
# Initialize mask shape of grid to true boolean (1 = True = Inclusion)
true_mask = np.ones_like(uit, dtype=bool)

# Set NaN indices to False (Exclusion)
true_mask[inan] = False

# Filter Ice Motion to true indices
uit_f = uit[true_mask]
vit_f = vit[true_mask]

# Take Wind, Concentration, Uncertainty, time at same indices
uwt_f = uwt[true_mask]
vwt_f = vwt[true_mask]

icy_f = icy[true_mask]

rt_f = rt[true_mask]

tt_f = tt[true_mask]
```

```python
def normalize(data):
    
    data_mean = np.nanmean(data)
    data_std = np.nanstd(data)
    data_norm = (data - data_mean)/data_std
    
    return data_norm
```

```python
# Normalize data

uit_n = normalize(uit_f)
vit_n = normalize(vit_f)
uwt_n = normalize(uwt_f)
vwt_n = normalize(vwt_f)

icy_n = normalize(icy_f)

rt_n = normalize(rt_f)
```

```python
# Convert to complex; 1j is python imaginary operator

it_c = uit_n + vit_n*1j # Complex Ice Motion Today
wt_c = uwt_n + vwt_n*1j # Complex Wind Today
icy_c = icy_n + 0*1j # Complex Previous Day Concentration (imag comp set to 0)
```

### Unweighted LR 

```python
# Define gram matrix
G = np.ones((len(it_c), 3), dtype = complex) # first column constant (1)

G[:,1] = wt_c # Complex wind, today
G[:,2] = icy_c # Complex ice concentration, yesterday

# Define data matrix
d = it_c.T

# Find model paremeters
m = (np.linalg.inv(G.T @ G)) @ G.T @ d

print(f"Model coeffeicients are: {np.array2string(m, precision=2, separator=', ')}")

# Calculate fit
fit = G @ m

# Add nan's back to fit
fit_full = np.full_like(uit, np.nan, dtype = complex)
fit_full[true_mask] = fit

u_fit = np.real(fit_full)
v_fit = np.imag(fit_full)
```

```python
# Normalize true values again for plotting

# Normalize data

uitnn = normalize(uit)
vitnn = normalize(vit)
uwtnn = normalize(uwt)
vwtnn = normalize(vwt)

icynn = normalize(icy)

rtnn = normalize(rt)
```

```python
# Plot last year of fit

plt.figure(figsize = (12,4))

plt.plot(tt[-365:], u_fit[-365:], label = 'Fit')
plt.plot(tt[-365:], uitnn[-365:], label = 'True')
plt.legend()
plt.title('u')
plt.show()

plt.figure(figsize = (12,4))

plt.plot(tt[-365:], v_fit[-365:], label = 'Fit')
plt.plot(tt[-365:], vitnn[-365:], label = 'True')
plt.legend()
plt.title('v')
plt.show()
```

```python
# Weighted and unweighted misfit
fit = u_fit
true = uitnn

r_1 = rtnn + 0.001

skill = 1 - np.nanvar(true - fit) / np.nanvar(true)
print(f'Zonal Skill is: {skill:.4f}')

skill_w = 1 - np.nanvar((true - fit) / r_1) / np.nanvar(true / r_1)
print(f'Zonal Weighted Skill is: {skill_w:.4f}')

mis = np.sqrt(((fit - true)) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis[-365:])
plt.title('Zonal Misfit, Unweighted')
plt.show()

mis_w = np.sqrt(((fit - true) / r_1) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis_w[-365:])
plt.title('Zonal Misfit, Weighted')
plt.show()

print('')

# Weighted and unweighted misfit
fit = v_fit
true = vitnn

skill = 1 - np.nanvar(true - fit) / np.nanvar(true)
print(f'Meridional Skill is: {skill:.4f}')

skill_w = 1 - np.nanvar((true - fit) / r_1) / np.nanvar(true / r_1)
print(f'Meridional Weighted Skill is: {skill_w:.4f}')

mis = np.sqrt(((fit - true)) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis[-365:])
plt.title('Meridional Misfit, Unweighted')
plt.show()

mis_w = np.sqrt(((fit - true) / r_1) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis_w[-365:])
plt.title('Meridional Misfit, Weighted')
plt.show()
```

```python
print(np.shape(rtnn))
print(np.shape(r_1))
```

### Weighted LR

```python
# Complex uncertainty (real), add 0.001 do avoid singular matrix

# 1 / r**2
# Complex conj

# NOTE - 3/14

# each value should be it's value multilplied by complex conjugate

# NOTE: DOES THIS COME FROM THE COVARIANCE?
# From Sarah = we have values in the form a + ib, if we have uncertainty for u and v, 
# we multipliy them together, but here we're multiplying the complex conjugate
# Here - the uncertainty is the same for u and v, so we have;

# --> B/c: (a**2 + ib**2) * (a**2 - ib**2) = (a ** 2 + b ** 2)
# if a = b
# --> 2 * (a ** 2)

# Add 0.001 to all to avoid divide by 0

rr = 2 * ((rt_n ** 2) + 0.001)
rr = 1 / rr

# Create diagonal matrix of uncertainty
R = np.diag(rr)

print(np.shape(R))

# Weighted model parameters
m_w = (LA.inv(G.T @ R @ G)) @ G.T @ R @ d # (eqn 39, SIOC221B Lec 10)

print(f'Weighted Model Parameters: {m_w}')

fit = G @ m_w

# Add nan's back to fit
fit_full = np.full_like(uit, np.nan, dtype = complex)
fit_full[true_mask] = fit

u_fit = np.real(fit_full)
v_fit = np.imag(fit_full)
```

```python
# Plot last year of fit

plt.figure(figsize = (12,4))

plt.plot(tt[-365:], u_fit[-365:], label = 'Fit')
plt.plot(tt[-365:], uitnn[-365:], label = 'True')
plt.legend()
plt.title('u')
plt.show()

plt.figure(figsize = (12,4))

plt.plot(tt[-365:], v_fit[-365:], label = 'Fit')
plt.plot(tt[-365:], vitnn[-365:], label = 'True')
plt.legend()
plt.title('v')
plt.show()
```

```python
# Weighted and unweighted misfit
fit = u_fit
true = uitnn

r_1 = rtnn + 0.001

skill = 1 - np.nanvar(true - fit) / np.nanvar(true)
print(f'Zonal Skill is: {skill:.4f}')

skill_w = 1 - np.nanvar((true - fit) / r_1) / np.nanvar(true / r_1)
print(f'Zonal Weighted Skill is: {skill_w:.4f}')

mis = np.sqrt(((fit - true)) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis[-365:])
plt.title('Zonal Misfit, Unweighted')
plt.show()

mis_w = np.sqrt(((fit - true) / r_1) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis_w[-365:])
plt.title('Zonal Misfit, Weighted')
plt.show()

print('')

# Weighted and unweighted misfit
fit = v_fit
true = vitnn

skill = 1 - np.nanvar(true - fit) / np.nanvar(true)
print(f'Meridional Skill is: {skill:.4f}')

skill_w = 1 - np.nanvar((true - fit) / r_1) / np.nanvar(true / r_1)
print(f'Meridional Weighted Skill is: {skill_w:.4f}')

mis = np.sqrt(((fit - true)) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis[-365:])
plt.title('Meridional Misfit, Unweighted')
plt.show()

mis_w = np.sqrt(((fit - true) / r_1) ** 2)

plt.figure(figsize = (12,4))
plt.plot(tt[-365:], mis_w[-365:])
plt.title('Meridional Misfit, Weighted')
plt.show()
```
