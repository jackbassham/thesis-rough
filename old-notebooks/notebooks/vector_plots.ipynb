```python
import cartopy.crs as ccrs
```

```python
import datetime

# Convert all time values to formatted strings (date only)
formatted_times = [
    datetime.datetime.strptime(str(t)[:10], "%Y-%m-%d").strftime("%b %d, %Y")
    for t in time
]
```

```python
import cmocean as cmo
```

```python
pres_idx = 1
prev_idx = pres_idx - 1

fig = plt.figure(figsize=(14, 4.5))

title = [f'True Meridional Ice Velocity, {formatted_times[pres_idx]}']
plt_data = vi[pres_idx,:,:]

cmap = 'RdBu'

# Left column: u values
ax_u = fig.add_subplot(projection=ccrs.SouthPolarStereo())
ax_u.set_extent([-180, 180, -90, -65], ccrs.PlateCarree())
pcm_u = ax_u.pcolormesh(lon, lat, plt_data, transform=ccrs.PlateCarree(), cmap=cmap)
ax_u.coastlines()
plt.colorbar(pcm_u, ax=ax_u, orientation='vertical', label = 'cm/s')
ax_u.set_title(title, fontweight='bold')

plt.show()
```
