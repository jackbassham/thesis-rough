```python
def norm_inputs(data):
    
    data_mean = np.mean(data)
    data_std = np.std(data)
    data_norm = (data - data_mean)/data_std

    return data_norm
```

```python
def calc_r2(fit, Y):
    """Calculates r^2 for u, v components given true 'Y' and predicted 'fit' complex values"""
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)

    # u componenet
    ru = (ufit.flatten() - uY.T) # Residuals
    SSRu = np.sum(ru**2) # Sum of square residuals
    SSTu = np.sum((uY.T - np.mean(uY.T))**2) # Total sum of squares, mean should be zero
    r2u = (1-(SSRu/SSTu)) # R^2 value

    # v componenet
    rv = (vfit.flatten() - vY.T)
    SSRv = np.sum(rv**2)
    SSTv = np.sum((vY.T - np.mean(vY.T))**2) 
    r2v = (1-(SSRv/SSTv))

    return r2u, r2v
```

```python
def calc_corr(fit, Y):
    """Calculates Pearson Correlation given true 'Y' and predicted 'fit' complex values"""
    # Phasing (no magnitude)
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)

    # Perason Correlation u
    corr1u = np.sum((uY-np.mean(uY))*(ufit-np.mean(ufit)))
    corr2u = np.sqrt(np.sum((uY-np.mean(uY))**2))*np.sqrt(np.sum((ufit-np.mean(ufit))**2))
    corru = corr1u/corr2u

    # Perason Correlation v
    corr1v = np.sum((vY-np.mean(vY))*(vfit-np.mean(vfit)))
    corr2v = np.sqrt(np.sum((vY-np.mean(vY))**2))*np.sqrt(np.sum((vfit-np.mean(vfit))**2))
    corrv = corr1v/corr2v

    return corru, corrv
```

```python
def new_calc_skill(pre, tru):
    """
    Calculates model skill represending the model error compared to the standard deviation of data
    
    Second term numerator is the Mean Squared Error
    
    Second term denominator is the variance in the true values
    
    """    
    
    skill1 = np.nanmean((tru - pre)**2, axis = 0) # MSE
    skill2 = np.nanmean((tru - np.nanmean(tru, axis = 0))**2, axis = 0) # Variance in true values
    
    if np.isnan(skill1).any() or np.isnan(skill2).any() or (skill2 == 0).any:
        print("Nan in skill")
    
    skill = 1 - skill1/skill2
    
    return skill
```

```python
def calc_skill(fit, Y):
    """
    
    Calculates model skill given true given true 'Y' and predicted 'fit' complex values
    
    Lauren et. al
    
    """
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)
    
    # Skill u component
    skill1u = np.sqrt(np.mean((ufit-uY)**2))
    skill2u = np.sqrt(np.mean((uY-np.mean(uY))**2))
    skillu = 1 - skill1u/skill2u

    # Skill v component
    skill1v = np.sqrt(np.mean((vfit-vY)**2))
    skill2v = np.sqrt(np.mean((vY-np.mean(vY))**2))
    skillv = 1 - skill1v/skill2v

    return skillu, skillv

## NEW SKILL - Weighted Skill##
    # Weighting points with large uncertainties less
# 1 - var( (m-itry)/uncertainty )/var(itry/uncertainty)
```

```python

# Mask ice concentration

ic_raw = np.round(ic * 250)

# Mask 
# 251 pole hole
# 252 unused data
# 253 coastline
# 254 land
ic = np.where((ic_raw == 251) | (ic_raw == 252) | (ic_raw == 253) | (ic_raw == 254), np.nan, ic)

# Filter points with ice error greater than 999 and 0 concentration ice motion
# Create mask
mask = (ic == 0) | (np.isnan(ic)) | (error >= 999) | (error <= -999)

# Mask points
ui = np.where(mask, np.nan, ui) 
uv = np.where(mask, np.nan, vi)
```

```python
# Shift present day parameters forward one day
ui_today = ui[1:,:,:]
vi_today = vi[1:,:,:]
uw_today = uw[1:,:,:]
vw_today = vw[1:,:,:]
time_today = time[1:]
error_today = error[1:,:,:]

# # Remove last day from previous day parameters
# ui_yester = ui[-1:,:,:]
# vi_yester = vi[-1:,:,:]
ic_yester = ic[:-1,:,:]
time_yester = time[:-1]

print("Days shifted")
```

```python
# Create dimensions for arrays based on lat lon grid
dims = np.shape(ui_today)
nt = dims[0]
ny = dims[1]
nx = dims[2]
dim = (ny, nx)

# Initialize fit array
u_true = np.zeros((nt, ny, nx))
v_true = np.zeros((nt, ny, nx))
u_pred = np.zeros((nt, ny, nx))
v_pred = np.zeros((nt, ny, nx))

# Initialize regression arrays
A = np.zeros((dim), dtype=np.complex64) # present day wind
C = np.zeros((dim), dtype=np.complex64) # prev day concentration

print("Arrays Initialized")
```

```python
# Iterate through lat and lon points
for iy in range(len(lat)):
    for ix in range(len(lon)):
# for iy in range(8):
#     for ix in range(20):
        # If all NaN values in motion vectors (NaN evaluetes to 'True' b/c not equal to 0)
        if np.all(np.isnan(ui_today[:,iy, ix]) | np.isnan(vi_today[:,iy,ix])):
            # Set arrays equal to NaN and continue to next iteration
            A[iy,ix] = np.nan
            C[iy,ix] = np.nan
            
            u_true[:,iy,ix] = np.nan
            v_true[:,iy,ix] = np.nan  
            u_pred[:,iy,ix] = np.nan
            v_pred[:,iy,ix] = np.nan
            
            continue

        else:
            try:          
                # Find NaN indices from ice motion set
                inan = np.where(np.isnan(ui_today[:,iy, ix]) | np.isnan(vi_today[:,iy,ix]))
                
                # Create NaN mask
                # Initialize mask shape of grid (initializes to true - inclusion)
                nmask = np.ones_like(ui_today[:,iy,ix], dtype=bool)
                # Set nan values to false
                nmask[inan] = False
                
                # Filter ice motion
                ui_to = ui_today[nmask,iy,ix]
                vi_to = vi_today[nmask,iy,ix]
                
                # Take Wind and Concentration at same filter indices
                uw_to = uw_today[nmask,iy,ix]
                vw_to = vw_today[nmask,iy,ix]
                
                ic_ye = ic_yester[nmask,iy,ix]
                
                
                # Filter dates non nan
                time_to = time_today[nmask]
        
                # Normalize data
                ui_to = norm_inputs(ui_to)
                vi_to = norm_inputs(vi_to)
                uw_to = norm_inputs(uw_to)
                vw_to = norm_inputs(vw_to)
                
                ic_ye = norm_inputs(ic_ye)
                
                # Save in true array
                u_true[nmask, iy, ix] = ui_to
                v_true[nmask, iy, ix] = vi_to
                
                u_true[inan, iy, ix] = np.nan
                v_true[inan, iy, ix] = np.nan
    
                # Convert to complex; 1j is python imaginary operator
                ivel = ui_to + vi_to*1j # Complex Present Day Ice Motion
                w = uw_to + vw_to*1j # Complex Present Day Wind
                ic = ic_ye + 0*1j # Complex Previous Day Concentration (imag comp set to 0)
    
                # Create input E matrix of Wind and Concentration (Transpose)
                E = np.array([w, ic]).T
    
                # Create y matrix of ice motion data points over time
                Y = np.atleast_2d(ivel).T
    
                # Initialize m vector of regression coefficients (sized number parameters)
                m = np.atleast_2d(np.array([np.zeros(2)])).T
    
                ### LR, No Ridge Parameter *(With Complex Conjugate Transpose)
                # Present Day Wind, Previous day concentration independent
                # Present Day Ice Motion, dependent
                m = (np.linalg.inv(E.conj().T@E))@E.conj().T@Y
                # m[:,iy,ix] = np.flatten(mtemp)
    
                # Extract Coefficients
                A[iy,ix] = m[0]
                C[iy,ix] = m[1]
    
                fit = E@m
        
#                 print(np.real(fit))
#                 print(np.imag(fit))
        
                u_pred[nmask, iy, ix] = np.real(fit).flatten()
                v_pred[nmask, iy, ix] = np.imag(fit).flatten()
            
                u_pred[inan, iy, ix] = np.nan
                v_pred[inan, iy, ix] = np.nan
    
#                 # Calculate statistics
#                 r2_u[iy,ix], r2_v[iy,ix] = calc_r2(fit_values, itry)
#                 skill_u[iy,ix], skill_v[iy,ix] = calc_corr(fit_values, itry)
#                 corr_u[iy,ix], corr_v[iy,ix] = calc_skill(fit_values, itry)
    
#             except np.linalg.LinAlgError:
#                 # If singular matrix set values to NaN
#                 A[iy, ix] = np.nan
#                 C[iy, ix] = np.nan
#                 r2_u[iy, ix] = np.nan
#                 r2_v[iy, ix] = np.nan
#                 skill_u[iy, ix] = np.nan
#                 skill_v[iy, ix] = np.nan
#                 corr_u[iy, ix] = np.nan
#                 corr_v[iy, ix] = np.nan
                
    
            except Exception as e:
                print(f"Error at iy={iy}, ix={ix}: {e}")

    print(f"lat: {iy} complete")
```
