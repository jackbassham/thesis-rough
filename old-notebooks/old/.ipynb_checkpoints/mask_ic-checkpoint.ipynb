```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
PATH_SOURCE = "/home/jbassham/jack/data/weddell/2020"
fnam = "con_nimbus7_latlon_sh_2020_2020.npz"
data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle=True)
ic = data['ic']
```

```python
plt.pcolormesh(ic[0,:,:])
plt.colorbar()
plt.show()
```

```python
ic_raw = np.round(ic * 250)
```

```python
plt.pcolormesh(ic_raw[0,:,:])
plt.colorbar()
plt.show()
```

```python
# Mask 
# 251 pole hole
# 252 unused data
# 253 coastline
# 254 land
ic_masked = np.where((np.round(ic*250 == 251)) | (np.round(ic*250 == 252)) | (np.round(ic*250 == 253)) | (np.round(ic*250 == 254)), np.nan, ic_raw)
```

```python
plt.pcolormesh(ic_masked[0,:,:])
```
